```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib widget
# Set the random seed for reproducibility
np.random.seed(10)

# Generate the data
x = 5 * np.random.random(100)
y = 2 + 4 * x + np.random.randn(100)

# Define the lines
re = 3 + 1.5 * x
re2 = 5 + 2 * x
re3 = 4 + 3 * x
re4 = 2 + 4 * x

# Create the figure and scatter plot
fig, ax = plt.subplots(figsize=(10, 5))
scatter = ax.scatter(x, y)
line_re, = ax.plot([], [], color='g', label='Intercept=3 & slope=1.5')
line_re2, = ax.plot([], [], color='b', label='Intercept=5 & slope=2')
line_re3, = ax.plot([], [], color='k', label='Intercept=4 & slope=3')
line_re4, = ax.plot([], [], color='r', label='Intercept=2 & slope=4')

# Setting up the plot
ax.set_xlim(0, 5)
ax.set_ylim(0, 25)
ax.legend()
ax.set_title('Fit Lines')

# Animation function
def animate(i):
    if i == 0:
        return scatter, line_re, line_re2, line_re3, line_re4  # Initialize all
    if i == 1:
        line_re.set_data(x, re)
    elif i == 2:
        line_re2.set_data(x, re2)
    elif i == 3:
        line_re3.set_data(x, re3)
    elif i == 4:
        line_re4.set_data(x, re4)
    return scatter, line_re, line_re2, line_re3, line_re4

# Create the animation
ani = FuncAnimation(fig, animate, frames=24, interval=1000, blit=True)
# ani.save('reg_animated.gif',writer='pillow',fps=1)
# Use 'save' before plt.show only if you want it displayed outside your code.
plt.show()
```
